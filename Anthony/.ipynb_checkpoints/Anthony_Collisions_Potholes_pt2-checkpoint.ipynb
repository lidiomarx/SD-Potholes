```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from config import gkey
import requests
```

```python
file = "../data/collision_output_pt1.csv"
df = pd.read_csv(file)
df.head()
```

```python
params = {"key": gkey}

print('Beginning Data Retrieval')

for index, row in df.iterrows():

    if pd.isnull(df.loc[index, 'zip_code']):
        
        print(f'Processing Record {index}')
        base_url = "https://maps.googleapis.com/maps/api/geocode/json"

        street_no = row['street_no']
        street_name = row['street_name']
        street_type = row['street_type']
        city = 'San Diego'
        state = 'CA'

        params['address'] = f"{street_no}{street_name}{street_type},{city},{state}"

        address_zips = requests.get(base_url, params=params)

        # print(address_zips.url)

        address_zips = address_zips.json()

        try:
            df.loc[index, "zip_code"] = address_zips['results'][0]['address_components'][7]['short_name']

        except (KeyError, IndexError):
            print('Zip Code not found. Skipping...')
    
df.head()
```

```python
df.to_csv("../data/collision_output_pt2.csv")
```
