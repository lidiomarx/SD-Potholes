```python
# Dependencies
import pandas as pd
import numpy as np
import requests
import json

# Google API Key
from config import gkey
```

```python
df = pd.read_csv("data/20190228_get_it_done_pothole_requests_datasd.csv")
df.head()
```

```python
#install pip module uszipcode
from uszipcode import SearchEngine

search = SearchEngine(simple_zipcode=True)

df['zipcode'] = ''

for index in df.index:
    if index % 100 == 0:
        print(index)
    lat = df.loc[index,'lat']
    long = df.loc[index,'long']
    
    try:
        result = search.by_coordinates(lat=lat, lng=long, radius=10, returns=1)
        zipcode = result[0].zipcode
        df.loc[index,'zipcode'] = zipcode
    except:
        print(f'Error:{index}')

```
