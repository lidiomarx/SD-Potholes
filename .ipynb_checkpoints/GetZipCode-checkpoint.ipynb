```python
# Dependencies
import pandas as pd
import numpy as np
import requests
import json
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./data/20190228_get_it_done_pothole_requests_datasd.csv")
df.head()
```

```python
#install pip module uszipcode
from uszipcode import SearchEngine

#use simple uszipcode module
search = SearchEngine(simple_zipcode=True)

#add a blank column name it zipcode
df['zipcode'] = ''
indices = []

#loop throught dataframe using lat & long, to find zipcode
for index in df.index:
    
#search for error in every 100 rows
    if index % 100 == 0:
        print(index)
    lat = df.loc[index,'lat']
    long = df.loc[index,'long']
    
#use module uszipcode search.by_coordinates to find zip code
    try:
        result = search.by_coordinates(lat=lat, lng=long, radius=30, returns=1)
        zipcode = result[0].zipcode
#if find zipcode, fill in the column under zipcode
        df.loc[index,'zipcode'] = zipcode
    except:
#if did not find zipcode, print out as error
        print(f'Error:{index}')
        indices.append(index)

```

```python
#show the zipcode column
df.head()
```

```python
#error row number, set as indices
indices
```

```python
#drop error rows
df = df.drop(df.index[indices])
df.head()
```

```python
df.to_csv('data/zipcode_20190228_get_it_done_pothole_requests_datasd.csv')
```

```python
# Dependencies
import pandas as pd
import numpy as np
import requests
import json
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/zipcode_20190228_get_it_done_pothole_requests_datasd.csv')
df.head()
```

```python
#value_count the number of frequency per zipcode, and then set it as DataFrame
df2=pd.DataFrame(data = df.zipcode.value_counts())
df2
```

```python
#reset DataFrame index
df3=df2.reset_index()
df3
```

```python
#find the top 10 zipcode as frequency
case_group2 = df3.nlargest(10, 'zipcode')
case_group2
```

```python
#rename column names
case_group1 = case_group2.rename(columns={'index':'zipcode', 'zipcode':'number of case'})
case_group1
```

```python
# define a bar chart from the data frame
case_group1.plot(kind='bar',x='zipcode', y='number of case')

#set new label name
labels = [92126, 92117, 92111, 92128, 92109, 92122, 92103, 92115, 92101, 92116]
fig = plt.figure(1)

# defining the x-axis tick labels
ax = fig.add_subplot(111)
ax.set_xticklabels(labels, rotation = 45)

# define plot title
plt.title('Top 10 Pothole frequency per ZipCode')

# defines the x axis & y axis title
plt.xlabel('Top 10 Zip Codes')
plt.ylabel('number of cases')
```
